```python
import json
import numpy as np
import pandas as pd
from urllib.error import URLError, HTTPError
from urllib.request import urlopen

class EIAgov(object):
    def __init__(self, token, series):
        '''
        Purpose:
        Initialise the EIAgov class by requesting:
        - EIA token
        - id code(s) of the series to be downloaded

        Parameters:
        - token: string
        - series: string or list of strings
        '''
        self.token = token
        self.series = series

    '''
    def __repr__(self):
        return str(self.series)
    '''

    def Raw(self, ser):
        # Construct url
        url = 'http://api.eia.gov/series/?api_key=' + self.token + '&series_id=' + ser.upper()

        try:
            # URL request, URL opener, read content
            response = urlopen(url);
            raw_byte = response.read()
            raw_string = str(raw_byte, 'utf-8-sig')
            jso = json.loads(raw_string)
            return jso

        except HTTPError as e:
            print('HTTP error type.')
            print('Error code: ', e.code)

        except URLError as e:
            print('URL type error.')
            print('Reason: ', e.reason)

    def GetData(self):
        # Deal with the date series                       
        date_ = self.Raw(self.series[0])        
        date_series = date_['series'][0]['data']
        endi = len(date_series) # or len(date_['series'][0]['data'])
        date = []
        for i in range (endi):
            date.append(date_series[i][0])

        # Create dataframe
        df = pd.DataFrame(data=date)
        df.columns = ['Date']

        # Deal with data
        lenj = len(self.series)
        for j in range (lenj):
            data_ = self.Raw(self.series[j])
            data_series = data_['series'][0]['data']
            data = []
            endk = len(date_series)         
            for k in range (endk):
                data.append(data_series[k][1])
            df[self.series[j]] = data
        
        return df

if __name__ == '__main__':
    tok = '24a8506ec725730b4b4c083a58f4f848'
        
    # Natural Gas - Daily prices
    # http://www.eia.gov/beta/api/qb.cfm?category=462457&sdid=NG.RNGC1.D
    #ng = states_ec #['EMISS.CO2-TOTV-TT-TO-AL.A','EMISS.CO2-TOTV-TT-TO-AK.A']#['ELEC.REV.AL-ALL.M', 'ELEC.REV.AK-ALL.M', 'ELEC.REV.CA-ALL.M']#['NG.RNGC1.D']  # w/ several series at a time 
    
    csC=states_co2C
    csP=states_co2P
    csI=states_co2I
    csR=states_co2R
    csT=states_co2T
    abr=states_abr
    ec=states_ec
    ep=states_ep
    rp=states_rp
    rg=states_rg
    sC=states_sC
    sP=states_sP
    sI=states_sI
    sR=states_sR
    sT=states_sT
    cd=states_cd
    data_co2C= EIAgov(tok, csC)
    data_co2P= EIAgov(tok, csP)
    data_co2I= EIAgov(tok, csI)
    data_co2R= EIAgov(tok, csR)
    data_co2T= EIAgov(tok, csT)
    data_abr=EIAgov(tok,abr)
    data_ec=EIAgov(tok,ec)
    data_ep=EIAgov(tok,ep)
    data_rp=EIAgov(tok,rp)
    data_rg=EIAgov(tok,rg)
    data_sC=EIAgov(tok,sC)
    data_sP=EIAgov(tok,sP)
    data_sI=EIAgov(tok,sI)
    data_sR=EIAgov(tok,sR)
    data_sT=EIAgov(tok,sT)
    data_cd=EIAgov(tok,cd)

    #print(data_ec.GetData())
```

```python
import requests
import lxml.html as lh
import pandas as pd
#url='https://en.wikipedia.org/wiki/List_of_countries_by_meat_consumption'
#url='https://en.wikipedia.org/wiki/Forest_cover_by_state_and_territory_in_the_United_States#cite_note-USDA1-1'
url='https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_area'
#url='https://www.enchantedlearning.com/usa/states/area.shtml'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
tr_elements = doc.xpath('//tr')
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[1]:
    i+=1
    name=t.text_content()
    print( '%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
#count=0
for j in range(1,len(tr_elements[0:3])):
    #print(j)
    #count+=1
    #T is our j'th row
    T=tr_elements[j]
    print(T)
```

```python
for t in tr_elements[61].iterchildren():
    print(t.text_content())
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(2,62):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=12:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```
